## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Line Item Insurance (USD) has missing values.Instead of deleting rows, filling missing insurance values with the median:
df['Line Item Insurance (USD)'] = df['Line Item Insurance (USD)'].fillna(
    df['Line Item Insurance (USD)'].median()
)

#replacing nulls of Dosage and shipment mode columns with "unknown"
df['Shipment Mode'] = df['Shipment Mode'].fillna('Unknown')
df['Dosage'] = df['Dosage'].fillna('Unknown')

#removing some more null values
df['PQ First Sent to Client Date'] = df['PQ First Sent to Client Date'].replace(['Pre-PQ Process','Date Not Captured','nan'], np.nan)

df['PO Sent to Vendor Date'] = df['PO Sent to Vendor Date'].replace(['Date Not Captured','N/A - From RDC'], np.nan)
df['Vendor INCO Term'] = df['Vendor INCO Term'].replace(['N/A - From RDC'], np.nan)

 #changing dtype to numeric
df['Freight Cost (USD)'] = pd.to_numeric(df['Freight Cost (USD)'], errors='coerce')
# dropping unnecessary col
df = df.drop(columns=['ID'])

# changing dtypes of columns having datetime format from object to datetime 
date_columns = [
    'Scheduled Delivery Date',
    'Delivered to Client Date',
    'Delivery Recorded Date'
]

for col in date_columns:
    df[col] = pd.to_datetime(df[col])

````

````python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Line Item Value'])
plt.show()
````

````python
numerical_cols = [
    'Unit of Measure (Per Pack)',
    'Line Item Quantity',
    'Line Item Value',
    'Pack Price',
    'Unit Price',
    'Weight (Kilograms)',
    'Freight Cost (USD)',
    'Line Item Insurance (USD)'
]
for col in numerical_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
outlier_summary = []

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    count = ((df[col] < lower) | (df[col] > upper)).sum()

    outlier_summary.append({
        'Column': col,
        'Outliers': count
    })

outlier_df = pd.DataFrame(outlier_summary)
print(outlier_df)
````

### What all manipulations have you done and insights you found?

Data wrangling is the process of cleaning and preparing the dataset before performing exploratory data analysis. The following preprocessing steps were carried out on the FedEx SCMS Delivery History dataset:

Handling Missing Values
The missing values in the Line Item Insurance (USD) column were replaced with the median value of the column to avoid losing valuable records and to minimize the effect of outliers.
Missing values in the Shipment Mode and Dosage columns were replaced with "Unknown" since these are categorical variables.
Cleaning Inconsistent Data
Some columns contained placeholder values such as "Pre-PQ Process", "Date Not Captured", and "N/A - From RDC" instead of actual missing values.
These placeholder values were replaced with NaN so that they could be correctly identified as missing data during analysis.
Data Type Conversion
The Freight Cost (USD) column was converted from object (string) type to numeric using pd.to_numeric() with errors='coerce', ensuring that invalid values were treated as missing values.
The date columns (Scheduled Delivery Date, Delivered to Client Date, and Delivery Recorded Date) were converted from object type to datetime format to enable date-based analysis.
Removing Unnecessary Columns
The ID column was removed because it is only a unique identifier and does not contribute to the analysis.
Final Clean Dataset
After completing these preprocessing steps, the dataset became clean, consistent, and suitable for exploratory data analysis and visualization.

#### Chart - 7 On time vs Late delivery rate

````python
# Chart - 7 visualization code
# no of days taken to deliver a product --> adding delivery time column to the data
df['Delivery Time'] = (df['Delivered to Client Date']-df['Scheduled Delivery Date']).dt.days

# 1 . caluculate on-time delivery rate
on_time_deliveries = df.loc[df['Delivery Time']<= 0]
on_time_delivery_rate = len(on_time_deliveries)*100/len(df)
late_deliveries = df.loc[df['Delivery Time']>0]
late_delivery_rate = len(late_deliveries)*100/len(df)

# PIE CHART FOR ONTIME AND LATE DELIVERY RATE
labels = ['On-Time', 'Late']
sizes = [on_time_delivery_rate, late_delivery_rate]
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('On-Time vs. Late Delivery Rate')
plt.show()
````

#### Chart - 8 On time vs Late delivery rate by Shipment mode

````python
# Chart - 8 visualization code
# Create delivery status
df['Delivery Status'] = np.where(
    df['Delivered to Client Date'] <= df['Scheduled Delivery Date'],
    'On Time',
    'Late'
)
#Calculate delivery rates by Shipment Mode
delivery_rate = (
    df.groupby('Shipment Mode')['Delivery Status']
      .value_counts(normalize=True)
      .mul(100)
      .unstack()
      .fillna(0)
      .round(2)
)

delivery_rate
#plotting the result
delivery_rate.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('On-Time vs Late Delivery Rate by Shipment Mode')
plt.ylabel('Percentage (%)')
plt.xlabel('Shipment Mode')
plt.legend(title='Delivery Status')
plt.xticks(rotation=45)
plt.show()
````

##### 1. Why did you pick the specific chart?

I picked a grouped bar chart because it compares on-time and late deliveries for each shipment mode.

##### 2. What is/are the insight(s) found from the chart?

The chart shows that some shipment modes have better on-time delivery performance than others.

##### 3. Will the gained insights help creating a positive business impact?


Yes. The company can choose shipment modes that have better delivery performance and work on improving the modes with more delays.
